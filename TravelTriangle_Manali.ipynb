```python
import numpy as np
package_data= []
duration_data = []
original_price_data = []
discount_percentage_data = []
city_data = []
hotel_rating_data = []
activity_data = []

for i in range(1,95):
    # Step 1: Identify the list
    url = f'https://traveltriangle.com/tour-packages/manali/page/{i}?destination_type=manali'
   # Step 2: Extract the HTML code
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"
    }
    response = requests.get(url, headers = headers)

    if response.status_code == 403:
        print("There's a 403 Error!")
        break
    html_code = response.text

    # Step 3 : Parse the HTML code to BeautifulSoup object
    soup = BeautifulSoup(html_code)
    containers = soup.find_all('div', attrs = {'class': 'col-12 p8 pt0'})
    for c in containers:
        # for each container scrape all data
        # Product title
        packages= c.find('span', attrs = {'class': 'fw9 m0 f16 pfc3'})
        if packages is None:
            package_data.append(np.nan)
        else:
            package_data.append(packages.text)
    # Product price
        durations = c.find('span', attrs = {'class': 'iblock sfc6'})
        if durations is None:
           duration_data.append(np.nan)
        else:
           duration_data.append(durations.text)

    # Customer rating and reviews
        original_prices = c.find('span', attrs = {'class': 'f12 pfc3 tdl ml8 at_oldprice fw4'})
        if original_prices is None:
            original_price_data.append(np.nan)
        else:
            original_price_data.append(original_prices.text)

    # Ratings
        discount_percentages = c.find('span', attrs = {'class': 'f12 fw4 pt2 pb2 pl8 pr8 radius20 pbc1 sfcw at_discount_label'})
        if discount_percentages is None:
           discount_percentage_data.append(np.nan)
        else:
            discount_percentage_data.append(discount_percentages.text)



    # cities
        cities = c.find('div', attrs = {'class': 'clearfix absolute t60'})
        if cities is None:
           city_data.append(np.nan)
        else:
           city_data.append(cities.text)


    ## ratings
        hotel_ratings = c.find("label", attrs = {"class" : "pr0 pt0 pl20"})
        if hotel_ratings is None:
           hotel_rating_data.append(np.nan)
        else:
           hotel_rating_data.append(hotel_ratings.text)

    ## activity
        activities = c.find('ul', attrs = {'class': 'package-tags at_package_tags'})
        if activities is None:
           activity_data.append(np.nan)
        else:
           activity_data.append(activities.text)
```

```python
df1 = pd.DataFrame({'Package':package_data,
                   'Package Duration':duration_data,
                   'Package Original Price':original_price_data,
                   'Package Discount %':discount_percentage_data,
                   'City':city_data,
                   'Hotel Rating':hotel_rating_data,
                   'Activity':activity_data})

df1.head()
```

```python
df['City'] = (df['City'].astype(str).str.replace('Cities:', '', regex=False).str.strip())
df['City'] = (df['City'].astype(str).str.replace(r'\(\d+D\)', '', regex=True).str.strip())
```

```python
df['Package Original Price'] = (
    df['Package Original Price']
    .astype(str)                           
    .str.replace('â¹', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace('/-', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.strip()
)
```

```python
df.columns
```

```python
df['Package Discount %'] = (df['Package Discount %'].astype(str).str.replace('Off', '', regex=False))
```

```python
df.columns
```

```python
df['Hotel Rating'] = (df['Hotel Rating'].astype(str).str.replace('Star', '', regex=False))
```

```python
df.columns
```

```python
df['Package Discount %'] = (df['Package Discount %'].astype(str).str.replace('%', '', regex=False))
```

```python
df[['Days', 'Nights']] = df['Package Duration'].str.extract(r'(\d+)\s*Days\s*&?\s*(\d+)\s*Nights')
df.head() 
```

```python
df.drop(columns=["Package Duration"],inplace=True)
```
